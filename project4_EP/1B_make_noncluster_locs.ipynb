```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
from multiprocessing import Pool
from astropy.coordinates import SkyCoord
import healpy as hp
import multiprocessing as mp
from tqdm import tqdm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d    
import random
from datetime import datetime

# Importing the dataset containing optical clusters
df = pd.read_csv('data/awd.csv', skiprows = 0)
pl143 = hp.read_map('data/COM_CompMap_Compton-SZMap-nilc-ymaps_2048_R2.00.fits')

def get_voids(iteration):
    pbar.update(1)

    ra = random.uniform(0,360)
    dec = random.uniform(-20, 80)
    # Convert the RA and DEC to galactic coordinates
    c = SkyCoord(ra,dec,frame='icrs',unit='deg')

    glat = c.galactic.b.degree
    glon = c.galactic.l.degree

    if (glat>-25) and (glat<25):
        return
    
    pix = hp.ang2pix(2048,glon,glat,lonlat=True,nest=False)
    if pl143[pix] < -0.6e-5:
        return np.array([ra,dec,glon,glat,pl143[pix]])
        

if __name__ == "__main__":                                                      
    iterations = 10000000
    pbar = tqdm(total=int(iterations/8), position=0, leave=True)
    with mp.Pool(processes=8) as pool:                                          
        df = np.array(pool.map(get_voids, range(iterations)))
        
vals = np.array([elem for elem in df if elem is not None])
np.savetxt('vals {}.txt'.format(str(datetime.now())), vals)        
```
